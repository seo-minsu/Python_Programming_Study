### Step 5. Create a pie chart presenting the male/female proportion
--> 성별의 비율을 보여주는 파이차트 생성

```python
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

proportions = [males, females]

plt.pie(
    proportions,
    labels = ['Males', 'Females'],
    shadow = False,
    colors = ['green', 'orange'],
    )
plt.axis('equal')
plt.title('Sex proportion')
plt.tight_layout()
plt.show()
```

```python
## males, females 할당
males = (titanic['Sex'] == 'male').sum()
females = (titanic['Sex'] == 'female').sum()

## 만드려는 차트에 리스트화
proportions = [males, females]

## 파이차트 생성(리스트한거, 라벨, 그림자, 색깔, )
## GPT : 파이 차트에 각 항목의 비율을 표시하려면 matplotlib의 plt.pie() 함수에서 autopct 매개변수를 사용하면 된다.
## 이 매개변수에 문자열 형식으로 표시할 퍼센트 형식을 지정할 수 있다
## 예를 들어, 소수점 없이 퍼센트를 나타내려면 '%1.0f%%'로 설정하고, 소수점 첫째 자리까지 표시하고 싶다면 '%1.1f%%'라고 입력하면 된다.

plt.pie(
    proportions,
    labels = ['Males', 'Females'],
    shadow = False,
    colors = ['green', 'orange'],
    explode = (0.15 , 0),  ## explode = (0.1, 0)  # 첫 번째 조각을 0.1만큼 분리
    startangle = 90,
    autopct = '%1.1f%%' ## GPT : 파이 차트에 각 항목의 비율을 표시하려면 matplotlib의 plt.pie() 함수에서 autopct 매개변수를 사용하면 된다. 이 매개변수에 문자열 형식으로 표시할 퍼센트 형식을 지정할 수 있다.
                                 ##소수점 없이 퍼센트를 나타내려면 '%1.0f%%'로 설정하고, 소수점 첫째 자리까지 표시하고 싶다면 '%1.1f%%'라고 입력하면 된다.
    )
plt.axis('equal')
plt.title('Sex proportion')
plt.tight_layout() ##그래프의 각 요소들이 겹치지 않고 보기좋게 표시되도록 자동으로 레이아웃 조정
plt.show()
```
