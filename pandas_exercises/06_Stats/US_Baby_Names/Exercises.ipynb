We are going to use a subset of [US Baby Names](https://www.kaggle.com/kaggle/us-baby-names) from Kaggle.  
In the file it will be names from 2004 until 2014


### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv).

### Step 3. Assign it to a variable called baby_names.

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv'
baby_names = pd.read_csv(url, sep=',')
```

### Step 4. See the first 10 entries

```python
baby_names.head(10)
```

### Step 5. Delete the column 'Unnamed: 0' and 'Id'

```python
## del baby_names['Unnamed: 0']  ##del 함수는 하나씩 삭제 가능
## del baby_names['Id']
baby_names.drop(['Unnamed: 0','Id'], axis=1)  ##두 개 이상 삭제 시 drop 사용
```

### Step 6. Is there more male or female names in the dataset?

```python
gender_counts = baby_names['Gender'].value_counts()
if gender_counts['M']>gender_counts['F']:
  print("Male are more")
else:
   print("Female are more")
```

```python
## 답지
baby_names['Gender'].value_counts()
```

### Step 7. Group the dataset by name and assign to names
## column 'Name' 으로 group --> names로 할당

```python
baby_names.drop(['Unnamed: 0','Id','Gender','State'],axis=1)
names = baby_names.groupby('Name').sum()
names.head()
print(names.shape)
names.sort_values("Count", ascending=0).head()

## 왜 답지랑 똑같이 안나올까....핳하ㅓㅎ
```

### Step 8. How many different names exist in the dataset?
## How many different --> nunique 함수 사용

```python
names.nunique()
```

```python
## 답지
len(names)
```

### Step 9. What is the name with most occurrences?

```python
## names.value_counts().idxmax() 이렇게 했을때 오류떴음. GPT는 groupby 하고 바로 valuecounts 못쓴다고 함.
## 오류의 원인은 names에 groupby()로 그룹화된 객체가 할당되어 있기 때문입니다. groupby()는 그룹화된 객체를 생성하지만, 바로 value_counts() 같은 함수는 사용할 수 없습니다.
## groupby() 객체에는 바로 사용할 수 있는 value_counts()가 없기 때문에 이 오류가 발생한 것입니다.
## --> GPT는 해결방법으로 size 함수를 알려주었으나, 이해가 잘 안가서 baby_names 로 찾음
baby_names['Name'].value_counts().idxmax()
```

```python
## 답지
names.Count.idxmax() ##group화된 객체에는 valuecounts가 아니고 Count 함수써야하는 듯
```

### Step 10. How many different names have the least occurrences?

```python
names.size().min()
```

```python
## names.Count.min() --> 이거 아님
```

```python
## 답지
len(names[names.Count == names.Count.min()])
```

### Step 11. What is the median name occurrence?

```python
names[names.Count == names.Count.median()]
```

### Step 12. What is the standard deviation of names?

```python
names.Count.std()
```
