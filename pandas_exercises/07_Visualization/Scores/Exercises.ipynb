### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Create the DataFrame that should look like the one below.

```python
data = {'first_name': ['Jason','Molly','Tina','Jake','Amy'], 'last_name':['Miller','Jacobson','Ali','MIlner','Cooze'], 'age': [42,52,36,24,73],'female':[0,1,1,0,1], 'preTestScore': [4,24,31,2,3], 'postTestScore':[25,94,57,62,70]}
dataframe = pd.DataFrame(data)

print(dataframe)

## 나오기는 하는데 뭔가 맨 처음 깔끔한 표 형식은 아님...ㅠ
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure(figsize=(10, 6))
plt.scatter(dataframe['preTestScore'], dataframe['postTestScore'], s=dataframe['age']*10, alpha=0.5)
plt.title('Scatterplot of Pre-Test and Post-Test Scores (Point size by Age)')
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')

for i in range(len(dataframe)):
    plt.annotate(dataframe['first_name'][i], (dataframe['preTestScore'][i], dataframe['postTestScore'][i]))

plt.grid(True)
plt.show()

##Step 1: preTestScore와 postTestScore 간의 산점도를 그리고, 각 점의 크기는 age 값에 10을 곱해서 나타냅니다.
##Step 2: 그래프 제목과 x축, y축 레이블을 추가합니다.
##Step 3: 각 점에 해당하는 이름(first_name)을 annotate() 함수를 이용해 점 바로 옆에 표시합니다.
##Step 4: 그래프를 표시합니다.
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

--> 이번에는 각 점의 크기를 postTestScore의 4.5배로 설정하고, 색상은 성별에 따라 결정되도록 하시오

```python
plt.figure(figsize=(10, 6))
colors = dataframe['female'].apply(lambda x: 'blue' if x == 0 else 'red')
plt.scatter(dataframe['preTestScore'], dataframe['postTestScore'],
            s=dataframe['postTestScore']*4.5,
            c=colors, alpha=0.6)

plt.title('Scatterplot of Pre-Test and Post-Test Scores (Size by Post-Test Score, Color by Sex)')
plt.xlabel('Pre-Test Score')
plt.ylabel('Post-Test Score')

plt.text(5, 90, 'Male (Blue)', color='blue', fontsize=12)
plt.text(5, 85, 'Female (Red)', color='red', fontsize=12)

plt.grid(True)
plt.show()

##점의 크기: postTestScore 값에 4.5를 곱한 값으로 크기를 설정합니다. 점이 클수록 해당 점수의 값이 높음을 나타냅니다.
##색상: 남성은 파란색 (blue), **여성은 빨간색 (red)**으로 표시됩니다.
```
